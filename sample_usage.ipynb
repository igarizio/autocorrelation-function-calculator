```python
import matplotlib.pyplot as plt
from ARMA import ARMA
from autocorrelation import autocorrelation_function
```

```python
phis = [0.1, 0.2, 0.3]
thetas = [0.1, 0.2, 0.5, 0.05]
n = 300
```

```python
arma = ARMA.generate_arma(n, phis, thetas, mu=0)
ar = ARMA.generate_ar(n, phis)
ma = ARMA.generate_ma(n, thetas, mu=0)
wn = ARMA.generate_wn(n)
```

```python
max_lags = 20

plt.title("ARMA")
plt.bar(range(max_lags), autocorrelation_function(arma, max_lags))
plt.show()

plt.title("AR")
plt.bar(range(max_lags), autocorrelation_function(ar, max_lags))
plt.show()

plt.title("MA")
plt.bar(range(max_lags), autocorrelation_function(ma, max_lags))
plt.show()

plt.title("White noise")
plt.bar(range(max_lags), autocorrelation_function(wn, max_lags))
plt.show()
```
